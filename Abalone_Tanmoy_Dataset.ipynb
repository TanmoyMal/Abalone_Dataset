```python
sns.pairplot(data[['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight','Viscera weight', 'Shell weight', 'Rings']])
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight','Viscera weight', 'Shell weight', 'Rings']])
plt.title('Boxplot of Features')
plt.show()
```

```python
avg_measurements = data[['Height', 'Whole weight', 'Shucked weight','Viscera weight', 'Shell weight', 'Rings']].mean()
sns.barplot(x=avg_measurements.index, y=avg_measurements.values)
plt.title('Barplot of Average Measurements')
plt.ylabel('Average Measurement')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Length'], y=data['Height'], hue=data['Sex'])
plt.title('Scatterplot of Length vs Height')
plt.xlabel('Length')
plt.ylabel('Height')
plt.legend(title='Species')
plt.show()
```

```python
corr=data[['Height', 'Whole weight', 'Shucked weight','Viscera weight', 'Shell weight', 'Rings']].corr()
sns.heatmap(corr)
```
